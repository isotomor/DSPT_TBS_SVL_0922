```python
data  = ["Hola buenas tardes", "Hola soy jose", "Hola pablo"]
```

Wordcloud no acepta una lista con palabras, si no un string muy grande. El se encarga de partir el string y contar la frecuencia de las palabras para representarlas

```python
data_join = ' '.join(data)
```

```python
# !pip install wordcloud
```

```python
from wordcloud import WordCloud


wordcloud = WordCloud().generate(data_join)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
