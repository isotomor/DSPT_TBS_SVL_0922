```python
from selenium import webdriver
import time
import pandas as pd

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

GUIA PARA ELEGIR COMO ESCRAPEAR COSAS

- Si el objeto tiene ID: find_element('id','__su_id__')
- Si el objeto tiene una clase que parece que solo usa el: find_element(By.CLASS_NAME,'__su_clase__')
- Si no: usamos el xpath.

Recordad: Para UN elemento concreto: find_element. Si hay más de uno: find_elements

Paso 1: Entrar en la web

```python
# especificamos el path hasta nuestro driver recién descargado:
chrome_driver_path = 'chromedriver.exe'
options  = webdriver.ChromeOptions()
```

```python
# Creamos el driver con el que nos vamos a manejar en la sesión de scrapeo:
driver = webdriver.Chrome(executable_path = chrome_driver_path, options = options)
```

Paso 3: Obtener los productos

El código de abajo es para encontrar la forma correcta de buscar los objetos. Lo idoneo es que mal tenga valor 0, significa que no hemos pasado por el except!

```python
#elementos = driver.find_elements(By.CLASS_NAME,'wt-grid__item-xs-6') #64/118
#elementos = driver.find_elements(By.TAG_NAME,'li') #63/1012

elementos = driver.find_elements(By.CLASS_NAME,'js-merch-stash-check-listing')

bien,mal = 0,0
for elemento in elementos:
    try:
        aux = elemento.find_element(By.CLASS_NAME,'v2-listing-card__title')
        #print(len(aux))
        bien += 1
    except:
        mal += 1
        pass

print(bien)
print(mal)
```

Ahora creamos el código para extraer los datos

```python
elementos = driver.find_elements(By.CLASS_NAME,'js-merch-stash-check-listing')

data = []
for elemento in elementos:
    titulo = elemento.find_element(By.CLASS_NAME,'v2-listing-card__title').text
    precio = elemento.find_elements(By.CLASS_NAME,"currency-value")[0].text
    seller = elemento.find_element(By.CLASS_NAME,'wt-mb-xs-1').find_elements(By.TAG_NAME,'span')[2].text
    # print(titulo)
    # print(precio)
    # print(seller)
    #break
    data.append({
        'nombre_producto':titulo,
        'precio':precio,
        'seller':seller,
        #'link':
    })

df = pd.DataFrame(data)
```
